```python
import cv2
from data import load_lena
import matplotlib.pyplot as plt
```

```python
img = load_lena()

h, w, _ = img.shape

rotating_point = (w//2, h//2) # center
rotating_point = (0, 0) # top left corner
rotating_point = (w, 0) # top right corner

angle = -10
scale = 0.90

rotation_matrix = cv2.getRotationMatrix2D(rotating_point, angle, scale)
rotated_image = cv2.warpAffine(img, rotation_matrix, (w,h))
```

```python
plt.subplot(121), plt.imshow(img), plt.title('Orig'), plt.grid(False)
plt.subplot(122), plt.imshow(rotated_image), plt.title('Rotated'), plt.grid(False)
plt.show()
```
