```python
import cv2

from data import load_lena
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = load_lena()

tx, ty = 100, 50 # Shift 100 pixels right and 50 pixels down
# tx, ty = -100, -50 # Shift 100 pixels left and 50 pixels up

translation_matrix = np.float32([
    [1, 0, tx],
    [0, 1, ty]
])

height, width, _ = img.shape

translated_image = cv2.warpAffine(img, translation_matrix, (width, height))
```

```python
plt.subplot(121), plt.imshow(img), plt.title('Orig'), plt.grid(False)
plt.subplot(122), plt.imshow(translated_image), plt.title('Translation example'), plt.grid(False)
plt.show()
```
