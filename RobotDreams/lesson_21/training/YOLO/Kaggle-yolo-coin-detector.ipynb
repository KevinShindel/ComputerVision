```python
from ultralytics import YOLO
```

```python
model_version = "yolo26n.pt"
yaml_path = '/kaggle/input/datasets/sh1nd3l/yolo-coin-dataset/coin_yolo_split/data.yaml'
```

```python
model = YOLO(model_version)
```

```python
results = model.train(
    data=yaml_path,
    epochs=150,
    imgsz=640,
    batch=8,
    device=[0, 1],
    patience=25,
    cache=False,
    workers=2,
    amp=True,
    project="coin_detector",
    name="yolo26n_coin_multiobj_640",

    # Augmentations (controlled)
    mosaic=0.7,
    mixup=0.05,
    close_mosaic=15,

    degrees=15.0,
    scale=0.5,
    translate=0.10,
    shear=2.0,
    perspective=0.0005,

    hsv_h=0.015,
    hsv_s=0.6,
    hsv_v=0.35,
    fliplr=0.5,
    flipud=0.1,

    verbose=True
)
```
